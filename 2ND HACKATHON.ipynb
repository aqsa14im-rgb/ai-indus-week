```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Remove noise
gray = cv2.medianBlur(gray, 3)

# Apply threshold
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

print("Image Preprocessing Done")
```

```python
custom_config = r'--oem 3 --psm 6'
text = pytesseract.image_to_string(thresh, config=custom_config)

print("\n===== OCR OUTPUT =====\n")
print(text)
```

```python
def categorize(item):
    if item.lower() == "bread":
        return "Bakery"
    elif item.lower() == "milk":
        return "Dairy"
    elif item.lower() == "chips":
        return "Snacks"
    else:
        return "Others"

df["Category"] = df["Item"].apply(categorize)

print("\n===== CATEGORIZED DATA =====\n")
print(df)
```

```python
category_summary = df.groupby("Category")["Total"].sum().reset_index()

overall_total = df["Total"].sum()

category_summary["Percentage"] = (
    category_summary["Total"] / overall_total * 100
)

print("\n===== SPENDING SUMMARY =====\n")
print(category_summary)

print("\nTotal Spending: Rs", overall_total)
```

```python
highest = category_summary.loc[
    category_summary["Percentage"].idxmax()
]

print("\nHighest Spending Category:")
print(highest["Category"], "-", round(highest["Percentage"], 2), "%")
```
